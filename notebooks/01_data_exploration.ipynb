```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("Imports réussis!")
```

```python
# Charge les données
df = pd.read_csv('../data/raw/players_22.csv', low_memory=False)

print(f"Dataset chargé!")
print(f"Nombre de lignes: {df.shape[0]:,}")
print(f"Nombre de colonnes: {df.shape[1]}")
```

```python
# Analyser les valeurs manquantes
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing_Count': missing,
    'Missing_Percentage': missing_pct
}).sort_values('Missing_Percentage', ascending=False)

print("Top 20 colonnes avec valeurs manquantes:")
print(missing_df[missing_df['Missing_Count'] > 0].head(20))
```

```python
# Graphique des valeurs manquantes
import matplotlib.pyplot as plt

top_missing = missing_df[missing_df['Missing_Count'] > 0].head(15)

plt.figure(figsize=(12, 6))
plt.barh(top_missing.index, top_missing['Missing_Percentage'], color='coral')
plt.xlabel('Pourcentage de valeurs manquantes (%)')
plt.title('Top 15 colonnes avec valeurs manquantes')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Colonnes qu'on va utiliser pour l'analyse
colonnes_importantes = [
    'short_name', 'age', 'nationality_name', 'overall', 'potential',
    'club_name', 'league_name', 'player_positions', 'value_eur',
    'wage_eur', 'height_cm', 'weight_kg', 'pace', 'shooting', 
    'passing', 'dribbling', 'defending', 'physic'
]

# Créer un subset avec seulement ces colonnes
df_subset = df[colonnes_importantes].copy()

print(f"Nouveau dataset créé!")
print(f"Nombre de colonnes: {df_subset.shape[1]}")
print(f"Colonnes sélectionnées: {', '.join(colonnes_importantes)}")
```

```python
# Sauvegarder pour la prochaine étape (nettoyage)
df_subset.to_csv('../data/processed/players_subset.csv', index=False)
print("Subset sauvegardé dans data/processed/players_subset.csv")
```

```python
# Distribution des âges
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df_subset['age'], bins=30, color='steelblue', edgecolor='black', alpha=0.7)
plt.xlabel('Âge')
plt.ylabel('Nombre de joueurs')
plt.title('Distribution des âges')
plt.axvline(df_subset['age'].mean(), color='red', linestyle='--', 
            label=f'Moyenne: {df_subset["age"].mean():.1f} ans')
plt.legend()
plt.grid(alpha=0.3)

# Distribution des notes overall
plt.subplot(1, 2, 2)
plt.hist(df_subset['overall'], bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.xlabel('Overall')
plt.ylabel('Nombre de joueurs')
plt.title('Distribution des notes Overall')
plt.axvline(df_subset['overall'].mean(), color='red', linestyle='--',
            label=f'Moyenne: {df_subset["overall"].mean():.1f}')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
